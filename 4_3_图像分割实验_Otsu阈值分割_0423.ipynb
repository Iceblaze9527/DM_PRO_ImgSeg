```
import os
import cv2
from skimage import img_as_float
from skimage.filters import threshold_multiotsu
import numpy as np
import matplotlib.pyplot as plt
```

```
src_path = "samples/test/"
```

```
def draw(image_mat):
    plt.imshow(image_mat,cmap='Accent')
    plt.show()
    #save
    plt.close()
```

```
def threshold_seg(method='NONE', classes=4, clipLimit=5, tileGrid=1):
    for _, _, files in os.walk(src_path):
        files.sort()
        for name in files:
            img_path = src_path + name
            image = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)

            if (method=='CLAHE'):
                image = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)     
                lab_channels = cv2.split(image)
                clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize= (tileGrid,tileGrid))
                lab_channels[0] = clahe.apply(lab_channels[0])
                image_enhanced = cv2.cvtColor(cv2.merge(lab_channels), cv2.COLOR_LAB2RGB)
                image = cv2.GaussianBlur(image_enhanced,(5,5), sigmaX = 9)

            image_gray = img_as_float(cv2.cvtColor(image, cv2.COLOR_RGB2GRAY))
            thresholds = threshold_multiotsu(image_gray, classes = classes)
            regions = np.digitize(image_gray, bins=thresholds)
            draw(image)
            draw(regions)
```

```
threshold_seg(method='NONE', classes=4)
```

```
threshold_seg(method='CLAHE', classes=4, clipLimit=5, tileGrid=1)
```
